```python
import tkinter as tk

class chipOrg:
    def __init__(self):
        self.win = tk.Tk()
        self.win.geometry(f"{self.win.winfo_screenwidth() - 100}x{self.win.winfo_screenheight() - 100}")

        self.player = 2

        self.add_player = tk.Button(self.win,text = '>',command = self.add)
        self.sub_player = tk.Button(self.win,text = '<',command = self.sub,state = tk.DISABLED)
        self.player_num = tk.Label(self.win,text = f'Number of player: {self.player}')
        self.choose_name = tk.Button(self.win,text = 'Next',command = self.choose_name)
        
        self.player_num.grid(row=0,column=0)
        self.sub_player.grid(row=0,column=1)
        self.choose_name.grid(row=0,column=2)
        self.add_player.grid(row=0,column=3)
        self.win.mainloop()
        
    def add(self):
        self.player += 1
        if self.player == 10:
            self.add_player['state'] = tk.DISABLED
        if self.player == 3:
            self.sub_player['state'] = tk.NORMAL
        self.player_num['text'] = f'Number of player: {self.player}'
        
    def sub(self):
        self.player -= 1
        if self.player == 2:
            self.sub_player['state'] = tk.DISABLED
        if self.player == 9:
            self.add_player['state'] = tk.NORMAL
        self.player_num['text'] = f'Number of player: {self.player}'
        
    def get_info(self):
        if self.name_entry.get() == '':
            name = f'Player {len(self.safe) + 1}'
        else:
            name = self.name_entry.get()
        self.name_entry.delete(0, tk.END)
        
        if self.price_entry.get() == '':
            price = 2000
        else:
            price = self.price_entry.get()
        self.price_entry.delete(0, tk.END)
        
        
        self.safe[name] = [int(price),0 - int(price)]
        self.ask_n['text'] = f'What would you like to name the player {len(self.safe) + 1}: '
        
        if len(self.safe) == self.player:
            self.ask_n.destroy()
            self.ask_p.destroy()
            self.name_entry.destroy()
            self.price_entry.destroy()
            self.okay.destroy()
            
            self.gl = tk.Label(self.win,text = 'Good Luck, Gentlemen',font=("Times New Roman", 18))
            self.gl.grid()
            self.start_button = tk.Button(self.win,text = 'Start a game', command = self.start)
            self.start_button.grid() 

        
    def choose_name(self):
        self.add_player.destroy()
        self.sub_player.destroy()
        self.choose_name.destroy()
        self.player_num.destroy()
        
        self.safe = {}
        
        self.name_entry = tk.Entry(self.win)
        self.price_entry = tk.Entry(self.win)
        self.ask_n = tk.Label(self.win,text = 'What would you like to name the player 1: ')
        self.ask_p = tk.Label(self.win,text = f'How much would you like to bet: ')
        self.okay = tk.Button(self.win,text = '✔︎',command = self.get_info)

        self.ask_n.grid(row=0,column=0)
        self.name_entry.grid(row=0,column=1)
        self.ask_p.grid(row=1,column=0)
        self.price_entry.grid(row=1,column=1)
        self.okay.grid(row=2,column=0)
        

        
    def start(self):
        self.start_button.destroy()
        self.gl.destroy()
        
        self.control = tk.Frame(self.win)
        self.players = tk.Frame(self.win)
        self.decisions = tk.Frame(self.win)
        self.control.grid(row=0,column=0)
        self.players.grid(row=2,column=0)
        self.decisions.grid(row=3,column=0)
        
        self.pot = 0
        self.pot_show = tk.Label(self.win, text = str(self.pot), font=("Times New Roman", 60))
        self.pot_show.grid(row=1,column=0)
        
        
        
        self.new_round = tk.Button(self.control,text='New Round',command = self.siva,state = tk.DISABLED)
        self.sd = tk.Button(self.control,text='Show Down',command = self.showdown,state = tk.DISABLED)
        self.new_round.grid(row=0,column=0) 
        self.sd.grid(row=0,column=1)
        

        
        self.nameLabel = []
        self.chips = []
        self.betnums = []
        self.index = 0
        for name in self.safe:
            self.name = tk.Label(self.players,text = name)
            self.chip = tk.Label(self.players,text = self.safe[name][0])
            self.betnum = tk.Label(self.players,text='0')
            
            self.nameLabel.append(self.name)
            self.chips.append(self.chip)
            self.betnums.append(self.betnum)
                
            if self.index <= 8:
                self.name.grid(row = 0,column = self.index)
                self.chip.grid(row = 0,column = self.index+1)
                self.betnum.grid(row = 1,column=self.index, columnspan=2)
            else:
                self.name.grid(row = 2,column = self.index-10)
                self.chip.grid(row = 2,column = self.index-10+1)
                self.betnum.grid(row = 3,column=self.index-10, columnspan=2)
            self.index += 2
        
        self.call = tk.Button(self.decisions,text='call',command=self.hecall)
        self.fold = tk.Button(self.decisions,text='fold',command=self.hefold)
        self.raised = tk.Button(self.decisions,text='raise',command=self.heraised)
        self.raisePrice = tk.Entry(self.decisions,width=4)
        self.call.grid(row=0,column=0)
        self.fold.grid(row=0,column=1)
        self.raised.grid(row=0,column=2)
        self.raisePrice.grid(row=0,column=3)
        
    
        self.saveIndex = None
        self.whostart = 0                                  #Has to be 0 <= x <= (self.player-1)       
        self.nameLabel[self.whostart]['bg'] = '#FFC0CB'
        
        self.count = 0
        self.whoplayin = tuple(self.safe.keys())
        self.whosin = list(zip(self.whoplayin,self.nameLabel,self.betnums,self.chips))
        self.hwmch = [0 for i in range(len(self.whoplayin))]
        
        self.sb = 5
        self.bb = 10
        self.maxbet = self.bb
        
        self.winner = []
        self.money = 0
        self.winnerNum = 0
        
        self.grave = []
                
        self.info = tk.Frame(self.win)
        self.info.grid()
        self.info.place(relx=1.0, rely=0, anchor='ne')
        
        self.turn = ('Pre-flop','Flop','Turn','River')
        self.when = tk.Label(self.info,text=self.turn[0])
        self.blindshow = tk.Label(self.info,text = f'SB: {self.sb} BB: {self.bb}')
        self.round = 1
        self.roundShow = tk.Label(self.info,text = f'Round: {self.round}')
        self.roundShow.grid(row=0,column=0)
        self.when.grid(row=1,column=0)
        self.blindshow.grid(row=2,column=0)
        

        sb = (self.whostart + 1) % len(self.whoplayin)

        self.pot += self.sb
        self.pot_show['text'] = self.pot

        self.whosin[sb][3]['text'] = str(int(self.whosin[sb][3]['text']) - self.sb)    
        self.whosin[sb][2]['text'] = str(self.sb)


        bb = (self.whostart + 2) % len(self.whoplayin)

        self.pot += self.bb
        self.pot_show['text'] = self.pot

        self.whosin[bb][3]['text'] = str(int(self.whosin[bb][3]['text']) - self.bb)    
        self.whosin[bb][2]['text'] = str(self.bb)
        
        
        
        
    def hecall(self):
        for i in range(len(self.whosin)):
            if self.whosin[i][1]['bg'] == '#FFC0CB':
                who = self.whosin[i]
                break
        
        self.pot += self.maxbet - int(who[2]['text'])
        self.pot_show['text'] = self.pot
        
        who[3]['text'] = str(int(who[3]['text']) - (self.maxbet - int(who[2]['text'])))    
            
        who[2]['text'] = str(self.maxbet)

        self.count += 1
        self.check()
        
        
    def heraised(self):
        try:
            typed = int(self.raisePrice.get())
            for i in range(len(self.whosin)):
                if self.whosin[i][1]['bg'] == '#FFC0CB':
                    who = self.whosin[i]
                    break
            bet = typed + int(who[2]['text'])
            if (typed <= int(who[3]['text'])) and (bet > self.maxbet):
                self.maxbet = bet
                self.raisePrice.delete(0, tk.END)

                self.pot += typed
                self.pot_show['text'] = self.pot

                who[3]['text'] = str(int(who[3]['text']) - typed) 
                who[2]['text'] = str(self.maxbet)

                self.count = 1
                self.check()
        except:
                  pass
            

    def hefold(self):
        for i in range(len(self.whosin)):
            if self.whosin[i][1]['bg'] == '#FFC0CB':
                self.whosin[i][1]['bg'] = '#F0F0F0'
                self.whosin[i-1][1]['bg'] = '#FFC0CB'
                
                self.whosin[i][1]['fg'] = '#bdbdbd'
                self.whosin[i][2]['fg'] = '#bdbdbd'
                self.whosin[i][3]['fg'] = '#bdbdbd'
                
                self.whosin.pop(i)
                break
        
        if len(self.whosin) == 1:                   
            self.winner.append(self.whosin[0][0])
            self.whosin[0][1]['bg'] = '#000000'
            self.whosin[0][1]['fg'] = '#FFD700'
            
            for i in range(len(self.betnums)):
                num = int(self.betnums[i]['text'])
                self.hwmch[i] += num
            
            self.call['state'] = tk.DISABLED
            self.fold['state'] = tk.DISABLED
            self.raised['state'] = tk.DISABLED
            self.raisePrice['state'] = tk.DISABLED
            self.new_round['state'] = tk.NORMAL  
            
        else:
            if (i == self.whostart) and (i == len(self.whosin)):      
                if self.saveIndex == None:
                    self.saveIndex = self.whostart
                self.whostart = 0
                
            if i < self.whostart:                                      
                if self.saveIndex == None:
                    self.saveIndex = self.whostart
                self.whostart -= 1
            self.check()
        
        
    def check(self):
        if self.count == len(self.whosin):        
            self.count = 0
            for i in range(len(self.whosin)):
                if self.whosin[i][1]['bg'] == '#FFC0CB':
                    self.whosin[i][1]['bg'] = '#F0F0F0'
                    break
                    
            if self.when['text'] == 'River':       
                self.call['state'] = tk.DISABLED
                self.fold['state'] = tk.DISABLED
                self.raised['state'] = tk.DISABLED
                self.raisePrice['state'] = tk.DISABLED
                self.sd['state'] = tk.NORMAL
                
                for i in range(len(self.betnums)):
                    num = int(self.betnums[i]['text'])
                    self.hwmch[i] += num
                    self.betnums[i]['text'] = 0
                    

            else:                                  
                self.whosin[self.whostart][1]['bg'] = '#FFC0CB'
                
                for i in ('Pre-flop','Flop','Turn'):
                    if i == self.when['text']:
                        nxt = self.turn.index(i) + 1
                        self.when['text'] = self.turn[nxt]
                        break
                
                self.maxbet = 0
                
                for i in range(len(self.betnums)):
                    num = int(self.betnums[i]['text'])
                    self.hwmch[i] += num
                    self.betnums[i]['text'] = 0
                    
        else:                                
            for i in range(len(self.whosin)):
                if self.whosin[i][1]['bg'] == '#FFC0CB':
                    self.whosin[i][1]['bg'] = '#F0F0F0'
                    nxt = (i + 1)%len(self.whosin)
                    self.whosin[nxt][1]['bg'] = '#FFC0CB'
                    break
                    
    def showdown(self):
        self.sd['state'] = tk.DISABLED
        self.when['text'] = 'Showdown'
        
        self.bind = self.win.bind("<Button-1>",self.collector)
        
        for i in self.nameLabel:
            if i['fg'] != '#bdbdbd':
                i['bg'] = '#FFC0CB'
                i['fg'] = '#000000'

    def collector(self,event):
        if (event.widget in self.nameLabel) and (event.widget['fg'] != '#bdbdbd'):
            if event.widget['bg'] == '#FFC0CB':
                self.winner.append(event.widget['text'])
                event.widget['bg'] = '#000000'
                event.widget['fg'] = '#FFD700'
                
                self.winnerNum += 1
                self.new_round['state'] = tk.NORMAL
                
            elif event.widget['fg'] == '#FFD700':
                self.winner.remove(event.widget['text'])
                event.widget['bg'] = '#FFC0CB'
                event.widget['fg'] = '#000000'
                
                self.winnerNum -= 1
                if self.winnerNum == 0:
                    self.new_round['state'] = tk.DISABLED
                
            
            
    def siva(self):
        count = 0
        for i in self.safe:                                                
            self.safe[i][0] = self.safe[i][0] - self.hwmch[count]
            self.hwmch[count] = 0
            count += 1
        
        
        self.money = self.pot // len(self.winner)                              
        for i in self.winner:
            self.safe[i][0] = self.safe[i][0] + self.money
        
        
        self.winner.clear()
        self.money = 0
        self.winnerNum = 0        
        if len(self.whosin) != 1:
            self.win.unbind('<Button-1>', self.bind)
        
        
        self.pot = 0
        self.pot_show['text'] = self.pot
        self.call['state'] = tk.NORMAL
        self.fold['state'] = tk.NORMAL
        self.raised['state'] = tk.NORMAL
        self.raisePrice['state'] = tk.NORMAL
        self.new_round['state'] = tk.DISABLED 
        
        
        for i in range(len(self.nameLabel)):
            if self.nameLabel[i]['fg'] == '#bdbdbd':
                self.chips[i]['fg'] = '#000000'
                self.betnums[i]['fg'] = '#000000'
            self.nameLabel[i]['bg'] = '#F0F0F0'
            self.nameLabel[i]['fg'] = '#000000'
            self.betnums[i]['text'] = 0
        
        
        count = 0
        for i in self.safe:
            self.chips[count]['text'] = self.safe[i][0]
            count += 1        
        
        
        self.count = 0
        self.when['text'] = self.turn[0]
        self.maxbet = self.bb
        
        
        if self.saveIndex == None:
            self.saveIndex = self.whostart
        gap = 0
        for i in range(len(self.chips)):
            i -= gap
            if int(self.chips[i]['text']) <= 0:
                gap += 1
                nameLabel = self.nameLabel.pop(i)
                name = nameLabel['text']
                safe = self.safe.pop(name)
                chip = self.chips.pop(i)
                betnum = self.betnums.pop(i)
                self.hwmch.pop(i)
                
                temp = list(self.whoplayin)
                temp.pop(i)
                self.whoplayin = tuple(temp)
                
                nameLabel['bg'] = "#F0F0F0"
                nameLabel['fg'] = "#E0E0E0"
                chip['bg'] = "#F0F0F0"
                chip['fg'] = "#F0F0F0"
                betnum['bg'] = "#F0F0F0"
                betnum['fg'] = "#F0F0F0"
                
                dead = (name,i,safe,nameLabel,chip,betnum)
                self.grave.append(dead)
                self.player -= 1
                
                if i <= self.saveIndex:
                    self.saveIndex -= 1
        self.whostart = (self.saveIndex + 1) % len(self.whoplayin)
        self.saveIndex = None
        self.nameLabel[self.whostart]['bg'] = '#FFC0CB'
        
        
        self.whosin = list(zip(self.whoplayin,self.nameLabel,self.betnums,self.chips))
        
        
        if self.player == 1:
            for widget in self.win.winfo_children():
                if widget != self.pot_show:
                    widget.destroy()
            self.pot_show['text'] = 'THE END\n' + f'The champion of the night: {self.whoplayin[0]}'
            return None
        
        
        self.round += 1
        self.roundShow['text'] = f'Round: {self.round}'
        
        
        sb = (self.whostart + 1) % len(self.whoplayin)

        self.pot += self.sb
        self.pot_show['text'] = self.pot

        self.whosin[sb][3]['text'] = str(int(self.whosin[sb][3]['text']) - self.sb)    
        self.whosin[sb][2]['text'] = str(self.sb)


        bb = (self.whostart + 2) % len(self.whoplayin)

        self.pot += self.bb
        self.pot_show['text'] = self.pot

        self.whosin[bb][3]['text'] = str(int(self.whosin[bb][3]['text']) - self.bb)    
        self.whosin[bb][2]['text'] = str(self.bb)        



test = chipOrg()
```
